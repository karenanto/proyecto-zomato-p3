# 🍽️ Exploración de Datos y Prototipo ETL - Zomato

Este notebook documenta el proceso inicial de Análisis Exploratorio de Datos (EDA) y validación de reglas de negocio que posteriormente se automatizaron en el pipeline de producción (los scripts `.py`).

Cumple con la fase analítica previa de **Extracción y Transformación**, integrando validaciones de esquemas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuración visual
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Extracción (Carga Inicial)

```python
raw_path = "../data/raw/zomato_clean.csv"
if os.path.exists(raw_path):
    df = pd.read_csv(raw_path)
    print(f"Dataset cargado con {df.shape[0]} filas y {df.shape[1]} columnas.")
    display(df.head())
else:
    print("El archivo zomato_clean.csv no se encuentra en data/raw/")
```

## 2. Inspección y Limpieza (Transformación Básica)
Revisamos los tipos de datos y los valores nulos.

```python
if 'df' in locals():
    print(df.info())
    print("\nValores Nulos por columna:")
    print(df.isnull().sum())
```

```python
if 'df' in locals():
    # Limpieza estándar (Ejemplo de lo que hace transform.py)
    df = df.drop_duplicates()
    print(f"Total de filas después de eliminar duplicados: {df.shape[0]}")
```

## 3. Análisis Exploratorio (EDA)

### 3.1. Distribución de Calificaciones (Ratings)

```python
if 'df' in locals() and 'rate' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df['rate'], bins=20, kde=True, color='teal')
    plt.title("Distribución de Ratings de Restaurantes")
    plt.xlabel("Rating")
    plt.ylabel("Frecuencia")
    plt.show()
```

### 3.2. Relación entre Votos y Calificación

```python
if 'df' in locals() and 'votes' in df.columns and 'rate' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x='rate', y='votes', alpha=0.5, color='coral')
    plt.title("Rating vs Número de Votos")
    plt.xlabel("Rating")
    plt.ylabel("Votos")
    plt.show()
```

### 3.3. Correlaciones

```python
if 'df' in locals():
    numeric_cols = df.select_dtypes(include=[np.number])
    if not numeric_cols.empty:
        plt.figure(figsize=(8, 6))
        sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
        plt.title("Matriz de Correlación")
        plt.show()
```
